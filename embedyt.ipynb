# YouTube Video Embedder

Embeds a YouTube video inline in the notebook from a URL using `pytube` and `IPython.display`.

```
from IPython.display import YouTubeVideo
from pytube import YouTube
def embed_yt(yt_url, width=560, height=315):
    yt = YouTube(yt_url)
    return YouTubeVideo(yt.video_id, width=width, height=height)
```
